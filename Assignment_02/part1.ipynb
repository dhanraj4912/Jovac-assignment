```python
import numpy as np
import time
```

```python
# 1
arr1d = np.arange(1, 11)
arr2d = np.arange(1, 10).reshape(3, 3)
arr3d = np.random.rand(3, 5, 3)

print(arr1d.shape, arr1d.size, arr1d.dtype)
print(arr2d.shape, arr2d.size, arr2d.dtype)
print(arr3d.shape, arr3d.size, arr3d.dtype)
```

```python
data = np.array([10, 20, 30, 40, 50, 60, 70, 80, 90])

first_three = data[:3]
alternate_elements = data[::2]
reversed_array = data[::-1]

print(first_three)
print(alternate_elements)
print(reversed_array)
```

```python
A = np.random.randint(1, 21, 5)
B = np.random.randint(1, 21, 5)

add = A + B
subtract = A - B
multiply = A * B
divide = A / B
dot_product = np.dot(A, B)
mean_A = np.mean(A)
median_A = np.median(A)
std_A = np.std(A)
var_A = np.var(A)
max_B = np.max(B)
min_B = np.min(B)
max_idx_B = np.argmax(B)
min_idx_B = np.argmin(B)

print(add, subtract, multiply, divide)
print(dot_product)
print(mean_A, median_A, std_A, var_A)
print(max_B, min_B, max_idx_B, min_idx_B)
```

```python
arr = np.arange(1, 13)

reshaped_2d = arr.reshape(4, 3)
reshaped_3d = arr.reshape(2, 2, 3)
transposed_2d = reshaped_2d.T

print(reshaped_2d)
print(reshaped_3d)
print(transposed_2d, transposed_2d.shape)
```

```python
arr_random = np.random.randint(10, 51, 15)

greater_25 = arr_random[arr_random > 25]
arr_random[arr_random < 30] = 0
count_div_5 = np.sum(arr_random % 5 == 0)

print(greater_25)
print(arr_random)
print(count_div_5)
```

```python
linspace_array = np.linspace(0, 1, 10)
identity_matrix = np.eye(4)
random_array = np.random.randint(1, 101, 20)
sorted_array = np.sort(random_array)
largest_5 = sorted_array[-5:]

print(linspace_array)
print(identity_matrix)
print(sorted_array)
print(largest_5)
```

```python
large_array1 = np.random.rand(100, 100)
large_array2 = np.random.rand(100, 100)

start_time = time.time()

result_matrix = np.matmul(large_array1, large_array2)
determinant = np.linalg.det(result_matrix)
inverse = None
if determinant != 0:
    inverse = np.linalg.inv(result_matrix)

end_time = time.time()
time_taken = end_time - start_time

print(determinant)
print(inverse)
print(time_taken)
```
